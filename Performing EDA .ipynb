```python
# Count plotting in logarithmic scale

def uniplot(df,col,title,hue =None):
    
    sns.set_style('whitegrid')
    sns.set_context('paper')
    plt.rcParams["axes.labelsize"] = 20
    plt.rcParams['axes.titlesize'] = 22
    plt.rcParams['axes.titlepad'] = 30
    
    
    temp = pd.Series(data = hue)
    fig, ax = plt.subplots()
    width = len(df[col].unique()) + 7 + 4*len(temp.unique())
    fig.set_size_inches(width , 8)
    plt.xticks(rotation=45)
    plt.yscale('log')
    plt.title(title)
 
    ax = sns.countplot(data = appdata_df, x= col, order=appdata_df[col].value_counts().index,hue = hue,palette="viridis") 
        
    plt.show()

uniplot(Cantpay_df,col='INCOME_RANGE',title='Income ranges',hue='CODE_GENDER')
```

```python
def corrmap(data,title):
    plt.figure(figsize=(15, 10))
    plt.rcParams['axes.titlesize'] = 25
    plt.rcParams['axes.titlepad'] = 70

    sns.heatmap(data, cmap="YlGnBu",annot=False)

    plt.title(title)
    plt.yticks(rotation=0)
    plt.show()
 
corrmap(data=target0,title='Correlation - Target0')
```

### Univariate analysis

```python
def boxpl(data,col,title):
    sns.set_style('whitegrid')
    sns.set_context('talk')
    plt.rcParams["axes.labelsize"] = 20
    plt.rcParams['axes.titlesize'] = 22
    plt.rcParams['axes.titlepad'] = 30
    
    plt.title(title)
    plt.yscale('log')
    sns.boxplot(data =Other_df, x=col,orient='v')
    plt.show()
    
boxpl(data=Other_df,col='AMT_INCOME_TOTAL',title='Boxplt for income amount - T0')
```
